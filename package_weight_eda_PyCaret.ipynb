```python
import pandas as pd

from pycaret.regression import RegressionExperiment
```

```python
df = pd.read_csv('dataset-train.csv')
df_s = pd.read_csv('dataset-test.csv')
```

## Train on smaller dataset df_s

```python
re_s = RegressionExperiment()
re_s.setup(df_s, target='GrossActualKg', session_id=123, max_encoding_ohe=-1)
```

```python
best_s = re_s.compare_models()
```

```python
re_s.plot_model(best_s, plot='feature')
```

```python
best_s.get_params
```

```python
re_s.plot_model(best_s, plot='parameter')
```
